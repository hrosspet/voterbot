```python
import math

VOTING_DELAY = 30*60 - 60
# using normal sqrt instead of approx_sqrt
weight_fct = math.sqrt

def remove_hr1(post):
    changed_post = post.copy()
    for i, p in enumerate(post['active_votes']):
        if p['voter'] == 'hr1':
            changed_post['active_votes'] = post['active_votes'][:i] + post['active_votes'][i+1:]
            break
    
    return changed_post

def parse_voting_times(active_votes):
    for vote in active_votes:
        vote['time'] = datetime.strptime(vote['time'], '%Y-%m-%dT%H:%M:%S')

def recalculate_weights(active_votes, created):
    rshares_total = 0
    weight_total = 0
    weight_hr1 = 0
    
    for i, vote in enumerate(active_votes):
        
        w_before = weight_fct(rshares_total)
        
        delay = (vote['time'] - created).total_seconds()
        rshares_time_adjusted = vote['rshares'] * min(1, delay / (30*60))
        
        rshares_total += vote['rshares']
        
        w = weight_fct(rshares_total) - w_before
        
        total_weight += w
        
        if vote['voter'] == 'hr1':
            weight_hr1 = w
        
        print('oring weight: %f' % vote['weight'])
        vote['weight'] = w
        print('new weight: %f' % vote['weight'])
    
    return weight_hr1 / weight_total * rshares_total

def inject_hr1(post, rshares_hr1):
    created = post['created']
    voting_time = created + timedelta(seconds=VOTING_DELAY)
    
    for i, vote in enumerate(post['active_votes']):
        if vote['time'] > voting_time:
            hr_index = i
            hr_vote = {
                'rshares': rshares_hr1,
                'time': voting_time
            }
            post['active_votes'] = post['active_votes'][:i] + \
                                    [hr_vote] + \
                                    post['active_votes'][i+1:]
            break
    
    hr1_reward = recalculate_weights(post['active_votes'], created)
        
def get_share_hr1(post, rshares_hr1):
    changed_post = post.copy()
    parse_voting_times(changed_post['active_votes'])        
    changed_post['active_votes'] = sorted(changed_post['active_votes'], key=lambda x: x['time'])
    
    hr1_reward = inject_hr1(changed_post, rshares_hr1)
        
    return hr1_reward

def get_mean_100percent_vote_hr1_rshares(raw_posts):
    hr_rshares = [int(x['rshares'])
                  for y in raw_posts for x in y['active_votes'] if x['voter'] == 'hr1']
    
    voting_times = [x['time']
                  for y in raw_posts for x in y['active_votes'] if x['voter'] == 'hr1']
    
    time_period = (max(voting_times) - min(voting_times))
    time_period_days = time_period.total_seconds() / 3600 / 24
    number_full_votes = time_period_days * 10
    
    hr_rshares = np.array(hr_rshares).sum()
    
    # adjust for reverse auction
    hr_rshares = hr_rshares / 29 * 30
    
    return hr_rshares / number_full_votes, hr_rshares, time_period
```
